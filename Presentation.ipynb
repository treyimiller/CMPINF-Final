### Income
Data Example:

```python
income = pd.read_csv("assets/income.csv")
income.head(2)
```

```python
population = pd.read_csv("assets/population.csv")
population.head(2)
```

```python
income = pd.read_csv("assets/income.csv")
population = pd.read_csv("assets/population.csv")
count = 0

neighborhood = "Oakwood"

for line in income['Neighborhood']:
    if line == neighborhood:
        break
    count += 1

neighborhoodIncome = income.at[count, 'Estimate; Total:']

count = 0
for line in population['Community']:
    if line == neighborhood:
        break
    count += 1

neighborhoodPopulation = population.at[count, 'pop_total']
neighborhoodPopulation = neighborhoodPopulation.replace(",", "")

calc = int(neighborhoodIncome) / int(neighborhoodPopulation)

print(calc)

```
